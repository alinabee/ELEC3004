# Sampling Theorem
In this simple notebook, we will show why the Nyquist sampling theorem is necessary. See also chapter 2 of [Unpingco's book](https://search.library.uq.edu.au/primo-explore/fulldisplay?docid=61UQ_ALMA51151217360003131&context=L&vid=61UQ&search_scope=61UQ_All&isFrbr=true&tab=61uq_all&lang=en_US).

Imagine that we are measuring the output of a blackbox (purpose and construction unknown). Let's called this a function called 'blackBox'.

```python
import numpy
import math

f = 1.0  # Hz, signal frequency

def blackBox(sr):
    
    t = numpy.arange(-1,1+1/sr,1/sr) # sample interval, symmetric for convenience later
    x = numpy.sin(2*numpy.pi*f*t)
    
    return x, t
```

We will measure the output of the blackbox at a fixed interval sampling rate $s_r$

```python
sr = 5.0 # Hz, sampling rate (ie. >= 2*f) 
x, t = blackBox(sr)
```

We guess that it is producing a sine curve, so we draw a near perfectly sampled sine curve for comparison

```python
sr2 = 1000.0
t2 = numpy.arange(-1,1+1/sr2,1/sr2) # sample interval, symmetric for convenience later
y = numpy.sin(2*numpy.pi*f*t2)
```

```python
import matplotlib.pyplot as plt

plt.plot(t,x,'o-')
plt.plot(t2,y,'-')
plt.xlabel('Time',fontsize=18)
plt.ylabel('Amplitude',fontsize=18)
plt.show()
```
