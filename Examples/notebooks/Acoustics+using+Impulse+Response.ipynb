# Acoustics using Impulse Response
In this notebook, we will be exploring how linear time invariant systems can be characterisized by its impulse response and apply this is sounds effects for bathrooms, stadiums and halls.

Use [librosa.load](https://librosa.github.io/librosa/generated/librosa.core.load.html#librosa.core.load) to load an audio file into an audio array. Return both the audio array as well as the sample rate:

```python
import librosa
dataPath = 'E:/Dev/workspace/trunk/courses/ELEC3004/2018/Data/impulse_bathroom/' #change this to your data path
#load the impulse
fileName = dataPath+'Impulse1.wav'
x, sr = librosa.load(fileName)
```

If you receive an error with librosa.load, you may need to [install ffmpeg](https://librosa.github.io/librosa/install.html#ffmpeg).

Display the length of the audio array and sample rate:

```python
print(x.shape)
print(sr)
```

### Bathroom Sound effect
Let's look at synthesizing a bathroom sound effect from the impulse response of the impulse.

```python
#load the bathroom impulse
fileNameResponse = dataPath+'Bathroom1.wav'
response, sr2 = librosa.load(fileNameResponse)
```

```python
plt.figure(figsize=(8, 4))
librosa.display.waveplot(response, sr=sr2)
```

```python
ipd.Audio(fileNameResponse) # load a local WAV file
```

Now load the sound sample that was recorded outside the bathroom and make it sound like it was spoken within the bathroom by using the impulse response function. 

```python
#load the bathroom impulse
fileNameTalk = dataPath+'Talk1.wav'
speech, sr3 = librosa.load(fileNameTalk)
```

```python
plt.figure(figsize=(8, 4))
librosa.display.waveplot(speech, sr=sr3)
```

```python
ipd.Audio(fileNameTalk) # load a local WAV file
```

```python
#convolve the recorded sound with the impulse response to get sound effect
effect = np.convolve(speech, response)
outFileName = dataPath+'effect1.wav'
librosa.output.write_wav(outFileName, effect, sr3)
```

```python
plt.figure(figsize=(8, 4))
librosa.display.waveplot(effect, sr=sr3)
```

```python
ipd.Audio(outFileName) # load a local WAV file
```
